```python
#S3 Contoh Operasi
ds1 = pd.Series([4, 6, 7, 8])
ds2 = pd.Series([10, 13, 16, 19])
dsa = ds1 + ds2  #Operasi Jumlah
dsb = ds1 - ds2
dsc = ds1 * ds2
dsd = ds1 / ds2
print(dsa)
print(dsb)
print(dsc)
print(dsd)
```

```python
#S4 Konver Pandas Module Series ke Python List
import pandas as pd
ds = pd.Series([88, 99, 11, 33, 55])
print("Pandas Series dan Type")
print(ds)
print(type(ds))

print("Konver Pandas Series ke Python List")
print(ds.tolist())
print(type(ds.tolist()))
```

# <right><div style="font-family: Times New Roman; font-size: 30px; border-radius : 10px; background-color: #a4e5efff; color: #000000ff; padding: 12px; line-height: 0.7;">6. MEMFILTER DATA </div></center>


```python
#M1 Memilih Memilih Teks

import pandas as pd

# Load data
url = "https://raw.githubusercontent.com/Alam-A99/Pandas_Latih/main/penjualan.csv"
df = pd.read_csv(url)

# --- Data asli ---
print("=== Data Asli ===")
print(df)

print("\n=== 1. Filter nama persis 'Budi' ===")
print(df[df['nama'].str.contains('Budi', case=False)])

print("\n=== 2. Filter nama yang mengandung huruf 'a' (case-insensitive) ===")
print(df[df['nama'].str.contains('a', case=False)])

print("\n=== 3. Filter nama yang diawali huruf 'D' ===")
print(df[df['nama'].str.startswith('D')])

print("\n=== 4. Bersihkan spasi jika ada (antisipasi) ===")
df['nama'] = df['nama'].str.strip()
print(df[df['nama'].str.contains('Budi', case=False)])
```

```python
#M2 Abaikan Huruf Besar/Kecil
df[df['nama'].str.contains('in')]

import re
df[df['nama'].str.contains('in', flags=re.IGNORECASE)]
```

```python
#M3 Memilih Data yang Berada di List/Dataframe Lain

# menampilkan kolom februari pada file penjualan yang bernilai 130 dan 220

feb_penjualan = [130, 220]
df[df['feb'].isin(feb_penjualan)]
```

```python
#M4 Memilih berdasarkan Satu Kondisi

import pandas as pd
import re

#M4.1 Memilih Angka 

#1. sama dengan
df[df['jan'] == 200]
df[df.jan.eq(200)]

#2. Memilih Angka tidak sama dengan
df[df['jan'] != 200]
df[df.jan.ne(200)]
df[~(df['jan'] == 200)]

#3. Memilih Angka Lebih Dari

df[df['jan'] > 200]
df[df.jan.gt(200)]

#4. Memilih Lebih Dari atau Sama Dengan
df[df['jan'] >= 200]
df[df.jan.ge(200)]

#5. Memilih Kurang Dari atau Sama Dengan
df[df['jan'] < 200]
df[df.jan.lt(200)]

# 6. Memilih Kurang Dari atau Sama Dengan
df[df['jan'] <= 200]
df[df.jan.le(200)]
```

```python
#M1 Metode Apply untuk Menjalankan Fungsi Pada Setiap Baris pada Data Series

data = [ 3, 5, 7, 10, 11, 15, 16, 21, 26]
d_int = pd.Series(data)
def kecilbesar(num):
    if num <10:
        return "kecil"
    elif num >=10 and num <16:
        return "sedang"
    else:
        return "besar"

d_int.apply(kecilbesar)
```
