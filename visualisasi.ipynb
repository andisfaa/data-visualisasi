# <right><div style="font-family: Times New Roman; font-size: 30px; border-radius : 10px; background-color: #a4e5efff; color: #000000ff; padding: 12px; line-height: 0.7;">1. VISUALISASI DATA DENGAN MATPLOTLIB </div></center>


```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load data dari link
url = "https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/Data_Regresi.csv"
df = pd.read_csv(url)

# 2. Nama kolom sesuai dengan data
df.rename(columns={
    df.columns[0]: "TIME",
    df.columns[1]: "Promo_X1",
    df.columns[2]: "Karyawan_X2",
    df.columns[3]: "Jual_Y"
}, inplace=True)

print("Kolom setelah rename:", df.columns.tolist())

# 3. Visualisasi

# --- A. Plot Jual_Y terhadap TIME
plt.figure(figsize=(10, 5))
plt.plot(df["TIME"], df["Jual_Y"], marker='o', linestyle='-', color='blue')
plt.title("Penjualan (Y) vs Waktu (TIME)")
plt.xlabel("TIME")
plt.ylabel("Jual (Y)")
plt.grid(True)
plt.tight_layout()
plt.show()

# --- B. Scatter Promo_X1 vs Jual_Y
plt.figure(figsize=(8, 5))
plt.scatter(df["Promo_X1"], df["Jual_Y"], color='red')
plt.title("Pengaruh Promo terhadap Penjualan")
plt.xlabel("Promo (X1)")
plt.ylabel("Jual (Y)")
plt.grid(True)
plt.tight_layout()
plt.show()

# --- C. Scatter Karyawan_X2 vs Jual_Y
plt.figure(figsize=(8,5))
plt.scatter(df["Karyawan_X2"], df["Jual_Y"], color='green')
plt.title("Pengaruh Jumlah Karyawan terhadap Penjualan")
plt.xlabel("Karyawan (X2)")
plt.ylabel("Jual (Y)")
plt.grid(True)
plt.tight_layout()
plt.show()

# --- D. Multiple line plot: Promo_X1 & Karyawan_X2 vs TIME
plt.figure(figsize=(10,5))
plt.plot(df["TIME"], df["Promo_X1"], marker='s', linestyle='--', color='orange', label='Promo (X1)')
plt.plot(df["TIME"], df["Karyawan_X2"], marker='d', linestyle=':', color='purple', label='Karyawan (X2)')
plt.title("Promo dan Karyawan vs Waktu")
plt.xlabel("TIME")
plt.ylabel("Nilai")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

# <right><div style="font-family: Times New Roman; font-size: 30px; border-radius : 10px; background-color: #a4e5efff; color: #000000ff; padding: 12px; line-height: 0.7;">2. VISUALISASI DATA DENGAN PLOTLY </div></center>


```python
import pandas as pd
import plotly.express as px

# Load data
url = "https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/Data_Regresi.csv"
df = pd.read_csv(url)

print(df.head())   # cek 5 baris awal
print(df.columns)  # cek nama kolom yang tersedia

# Scatter plot: Promo(X1) vs Jual(Y)
fig = px.scatter(df, x="Promo(X1)", y="Jual(Y)", 
                 title="Scatter: Promo(X1) vs Jual(Y)",
                 labels={"Promo(X1)": "Promo (X1)", "Jual(Y)": "Penjualan (Y)"})
fig.show()

# Scatter plot: Karyawan(X2) vs Jual(Y)
fig2 = px.scatter(df, x="Karyawan(X2)", y="Jual(Y)", 
                  title="Scatter: Karyawan(X2) vs Jual(Y)",
                  labels={"Karyawan(X2)": "Jumlah Karyawan (X2)", "Jual(Y)": "Penjualan (Y)"})
fig2.show()

# Histogram distribusi penjualan
fig3 = px.histogram(df, x="Jual(Y)", nbins=20,
                    title="Distribusi Penjualan (Y)")
fig3.show()

# Line plot perkembangan penjualan dari waktu ke waktu
fig4 = px.line(df, x="TIME", y="Jual(Y)", 
               title="Tren Penjualan dari Waktu ke Waktu",
               labels={"TIME": "Waktu", "Jual(Y)": "Penjualan (Y)"})
fig4.show()
```

# <right><div style="font-family: Times New Roman; font-size: 30px; border-radius : 10px; background-color: #a4e5efff; color: #000000ff; padding: 12px; line-height: 0.7;">3. VISUALISASI DATA DENGAN SEABORN </div></center>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load data
url = "https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/Data_Regresi.csv"
df = pd.read_csv(url)

print(df.head())
print(df.columns)

# Step 2: Scatterplot Promo(X1) vs Jual(Y)
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="Promo(X1)", y="Jual(Y)", hue="TIME", palette="viridis")
plt.title("Scatter: Promo(X1) vs Jual(Y)")
plt.show()

# Step 3: Scatterplot Karyawan(X2) vs Jual(Y)
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="Karyawan(X2)", y="Jual(Y)", hue="TIME", palette="plasma")
plt.title("Scatter: Karyawan(X2) vs Jual(Y)")
plt.show()

# Step 4: Histogram distribusi Penjualan (Y)
plt.figure(figsize=(6,4))
sns.histplot(data=df, x="Jual(Y)", bins=20, kde=True, color="skyblue")
plt.title("Distribusi Penjualan (Y)")
plt.show()

# Step 5: Lineplot tren penjualan dari waktu ke waktu
plt.figure(figsize=(8,4))
sns.lineplot(data=df, x="TIME", y="Jual(Y)", marker="o")
plt.title("Tren Penjualan dari Waktu ke Waktu")
plt.show()

# Step 6: Pairplot untuk semua variabel numerik
sns.pairplot(df[["Promo(X1)", "Karyawan(X2)", "Jual(Y)"]], diag_kind="kde")
plt.suptitle("Pairplot antar Variabel", y=1.02)
plt.show()

# Step 7: Heatmap korelasi
plt.figure(figsize=(6,4))
sns.heatmap(df[["Promo(X1)", "Karyawan(X2)", "Jual(Y)"]].corr(), annot=True, cmap="coolwarm")
plt.title("Korelasi antar Variabel")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("https://raw.githubusercontent.com/Alam-A99/Case-Study/refs/heads/main/Data_Regresi.csv")

# Set theme Seaborn
sns.set_theme(style="ticks")

# Inisialisasi figure
f, ax = plt.subplots(figsize=(10, 6))

# Boxplot Jual(Y) vs Promo(X1)
sns.boxplot(
    data=df,
    x="Jual(Y)",
    y="Promo(X1)",      
    whis=[0, 100],      
    width=0.6,
    palette="RdPu",    
ax=ax
)

# Stripplot untuk tiap titik data
sns.stripplot(
    data=df,
    x="Jual(Y)",
    y="Promo(X1)",
    size=4,
    color=".3",        
    ax=ax
)

# Grid & styling
ax.xaxis.grid(True)
ax.set(ylabel="Promo(X1)")
sns.despine(trim=True, left=True)
plt.title("Boxplot & Stripplot: Penjualan vs Promo", fontsize=14, fontweight="bold")
plt.show()
```
